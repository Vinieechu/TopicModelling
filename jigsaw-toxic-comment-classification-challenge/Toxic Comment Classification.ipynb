```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
import matplotlib.pyplot as plt
%matplotlib inline
import re
import string
```

```python
#Null cleaning function

def myfillna(series):
    if series.dtype is pd.np.dtype(float):
        return series.fillna('')
    elif series.dtype is pd.np.dtype(int):
        return series.fillna('')
    else:
        return series.fillna('NA')
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
train=pd.read_csv('train.csv',encoding='latin-1')
```

```python
train=train.apply(myfillna)
```

```python
train.head()
```

```python
train.comment_text=train.comment_text.str.replace('[^a-zA-Z]', ' ') #replacing any character that is not a-z with a space
```

```python
#Remove unnecessary punctuations
train.comment_text=train['comment_text'].apply(lambda x:''.join([i for i in x if i not in string.punctuation]))
```

# Scattertext for analysis

```python
import scattertext as st
import spacy
import en_core_web_sm
```

```python
train.head()
```

```python
train['hate']=np.where(train.identity_hate==1,'Hate','Low_hate')
```

```python
x3=train[(train.toxic==0) &  (train.severe_toxic==0) & (train.obscene==0) & (train.threat==0) & (train.insult==0)& (train.identity_hate==0)].iloc[:10000]
```

```python
x2=train[train.identity_hate==1]
```

```python
x=x3.append(x2)
```

```python
nlp = en_core_web_sm.load()
corpus = st.CorpusFromPandas(x,category_col='hate',text_col='comment_text',nlp=st.whitespace_nlp_with_sentences).build()
term_freq_df = corpus.get_term_freq_df()
term_freq_df['Hate'] = corpus.get_scaled_f_scores('Hate')
term_freq_df['Low_hate'] = corpus.get_scaled_f_scores('Low_hate')
```

```python
term_freq_df.sort_values('Hate',ascending=False)
```

```python
html = st.produce_scattertext_explorer(corpus,category='Hate',category_name='High_hate',not_category_name='Low_hate',width_in_pixels=1000,metadata=x['hate'])
open("Hate_vs_low_hate.html", 'wb').write(html.encode('utf-8'))
```

```python
x['Formula']=np.where(x.Formula_Classifier==1,'Formula','UX_Bugs')
nlp = en_core_web_sm.load()
corpus = st.CorpusFromPandas(x,category_col='Formula',text_col='comment_text',nlp=st.whitespace_nlp_with_sentences).build()
```

```python
term_freq_df = corpus.get_term_freq_df()
term_freq_df['Formula'] = corpus.get_scaled_f_scores('Formula')
term_freq_df['UX_Bugs'] = corpus.get_scaled_f_scores('UX_Bugs')
html = st.produce_scattertext_explorer(corpus,category='Formula',category_name='Formula',not_category_name='UX_Bugs',width_in_pixels=1000,metadata=x['Formula'])
open("Formula_UX_Bug.html", 'wb').write(html.encode('utf-8'))
```
